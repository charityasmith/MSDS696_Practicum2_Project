### Notebook Description
This notebook handles initial data collection, scraping NBA statistics, and performing preliminary exploratory data analysis (EDA) to investigate whether the NBA‚Äôs increased emphasis on three-point shooting has positively or negatively affected the game.

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import PercentFormatter
import seaborn as sns
from scipy.stats import linregress
import os
```

```python
# Load all datasets
reg_season_per_game = pd.read_csv("../data/raw/reg_season_per_game_avg.csv")
reg_season_total = pd.read_csv("../data/raw/reg_season_total.csv")
playoff_per_game = pd.read_csv("../data/raw/playoff_per_game_avg.csv")
playoff_total = pd.read_csv("../data/raw/playoff_total.csv")
three_point_leaders=pd.read_csv("../data/raw/three_point_scoring_leaders.csv")
```

```python
print("üìã Column headers:")

print("\nRegular Season Per Game:")
print(reg_season_per_game.columns.tolist())

print("\nRegular Season Total:")
print(reg_season_total.columns.tolist())

print("\nPlayoff Per Game:")
print(playoff_per_game.columns.tolist())

print("\nPlayoff Total:")
print(playoff_total.columns.tolist())

print("\nThree-Point Leaders:")
print(three_point_leaders.columns.tolist())
```

```python
# Drop columns not relevant to 3P or scoring trends
cols_to_drop = [
    'FT', 'FTA', 'FT%', 'FT/FGA',
    'MP', 'PF',
    'TOV%', 'ORB%',
    'STL', 'BLK'
]

# Drop columns from all except three_point_leaders
reg_season_per_game = reg_season_per_game.drop(columns=cols_to_drop, errors='ignore')
reg_season_total = reg_season_total.drop(columns=cols_to_drop, errors='ignore')
playoff_per_game = playoff_per_game.drop(columns=cols_to_drop, errors='ignore')
playoff_total = playoff_total.drop(columns=cols_to_drop, errors='ignore')
```

```python
# Check for missing values
print("üîç Any missing values:")

print("Regular Season Per Game:", reg_season_per_game.isnull().values.any())
print("Regular Season Total:", reg_season_total.isnull().values.any())
print("Playoff Per Game:", playoff_per_game.isnull().values.any())
print("Playoff Total:", playoff_total.isnull().values.any())
print("Three-Point Leaders:", three_point_leaders.isnull().values.any())
```

```python
three_point_leaders.columns = three_point_leaders.columns.str.strip()
three_point_leaders = three_point_leaders.dropna(subset=['3-point Field Goals Made'])

# Convert to numeric if necessary
three_point_leaders['3-point Field Goals Made'] = pd.to_numeric(
    three_point_leaders['3-point Field Goals Made'], errors='coerce'
)
```

```python
# Summarize dataframes
def summarize(df, name):
    print(f"\nüìä Summary for {name}:")
    print(df.describe(include='all'))

summarize(reg_season_per_game, "Regular Season Per Game")
summarize(reg_season_total, "Regular Season Total")
summarize(playoff_per_game, "Playoff Per Game")
summarize(playoff_total, "Playoff Total")
summarize(three_point_leaders, "Three-Point Leaders")
```

### Correlation - Pace vs 3PA since 1995

```python
# Filter data to 1995 and on
modern_data = reg_season_total[reg_season_total['Season'] >= 1995]

# Compute correlation
slope, intercept, r_value, p_value, std_err = linregress(modern_data['Pace'], modern_data['3PA'])

# Choose darker green for strong correlation
trend_color = '#207f2f' if abs(r_value) >= 0.7 else '#ff7f0e' if abs(r_value) >= 0.4 else 'gray'

# Start plot
plt.figure(figsize=(10, 6))
sns.regplot(
    data=modern_data,
    x='Pace',
    y='3PA',
    scatter_kws={'alpha': 0.7, 's': 60, 'color': '#59a14f'},
    line_kws={'color': trend_color, 'linewidth': 2, 'alpha': 0.6},
    ci=None
)

# Highlight current season (latest year)
latest = modern_data[modern_data['Season'] == modern_data['Season'].max()]
plt.scatter(
    latest['Pace'],
    latest['3PA'],
    color='purple',
    s=80,
    edgecolor='black',
    zorder=5
)
plt.text(
    latest['Pace'].values[0],
    latest['3PA'].values[0],
    str(latest['Season'].values[0]),
    fontsize=10,
    fontweight='bold',
    color='purple',
    ha='right',
    va='bottom'
)

# Annotate 1999 (lockout), 2012 (lockout), 2020 (COVID)
annotate_years = [1999, 2012, 2020]
annot_colors = {'1999': 'brown', '2012': 'brown', '2020': 'brown'}

for yr in annotate_years:
    row = modern_data[modern_data['Season'] == yr]
    if not row.empty:
        x = row['Pace'].values[0]
        y = row['3PA'].values[0]
        plt.text(
            x, y,
            str(yr),
            fontsize=10,
            fontweight='bold',
            color=annot_colors[str(yr)],
            ha='center',
            va='bottom'
        )

# Labels & formatting
plt.title("Regular Season: Pace vs 3-Point Attempts", fontsize=14, weight='bold')
plt.xlabel("Possessions per Game", fontsize=12, weight='bold')
plt.ylabel("3-Point Attempts", fontsize=12, weight='bold')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

# Clean visuals
sns.despine()
plt.tick_params(axis='both', labelsize=10)

plt.tight_layout()

# Save the figure
plt.savefig("../visualizations/pace_vs_3pa.png", dpi=300)

# Display the plot
plt.show()
```

### Regular Season vs. Playoff 3P% - Difference Plot (Reg ‚Äì Playoff 3P%)

```python
# Select only the necessary columns for clarity
reg_3p = reg_season_total[['Season', '3P%']].rename(columns={'3P%': '3P%_Reg'})
playoff_3p = playoff_total[['Season', '3P%']].rename(columns={'3P%': '3P%_Playoff'})

# Merge on Season
merged = pd.merge(reg_3p, playoff_3p, on='Season', how='inner')

# Calculate the difference
merged['3P_Diff'] = merged['3P%_Reg'] - merged['3P%_Playoff']

# Create plot
plt.figure(figsize=(12, 6))
sns.lineplot(
    x=merged['Season'],
    y=merged['3P_Diff'],
    color='purple',
    linewidth=2  # no marker
)

# Reference line at 0
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# Shaded ¬±1% zone
plt.axhspan(-0.01, 0.01, color='gray', alpha=0.1)

# Titles and labels
plt.title("Difference in 3-Point Percentage", fontsize=15, weight='bold')
plt.xlabel("Season", fontsize=12, weight='bold')
plt.ylabel("% Difference (Regular Season - Playoffs)", fontsize=12, weight='bold')

# Format y-axis with ¬± percentage
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:+.1%}"))

# Style cleanup
sns.despine()
plt.tick_params(axis='both', labelsize=10)
plt.tight_layout()

# Save the figure
plt.savefig("../visualizations/diff_in_percentage.png", dpi=300)

# Show plot
plt.show()
```

```python
# Create plot with annotation for shortened 3 point line
plt.figure(figsize=(12, 6))
sns.lineplot(x=merged['Season'], y=merged['3P_Diff'], color='purple', linewidth=2)

# Reference line at 0
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# Optional shaded zone for small differences (~¬±1%)
plt.axhspan(-0.01, 0.01, color='gray', alpha=0.1)

# Highlight shortened 3-point line (1994‚Äì1997)
plt.axvspan(1994, 1997, color='orange', alpha=0.15, label='Shortened 3P Line (1994‚Äì1997)')

# Highlight Curry era (2015‚Äìpresent)
plt.axvspan(2015, reg_season_total['Season'].max(), color='purple', alpha=0.1, label='Curry Era (2015‚Äì)')

# Titles and labels
plt.title("Difference in 3-Point Percentage", fontsize=15, weight='bold')
plt.xlabel("Season", fontsize=12, weight='bold')
plt.ylabel("% Difference (Regular Season - Playoffs)", fontsize=12, weight='bold')

# Format y-axis with ¬± percentage
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:+.1%}"))

# Clean styling
sns.despine()
plt.tick_params(axis='both', labelsize=10)
plt.legend(loc='upper right')

plt.tight_layout()

# Save the figure
plt.savefig("../visualizations/diff_in_percentage_2.png", dpi=300)

# Display the plot
plt.show()
```

### Calculate Top 10 All-Time Leaders by Total 3PM

### Offensive Rating (how many points per 100 possessions) vs. 3PA

```python
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
import seaborn as sns

# Filter modern data
modern_data = reg_season_total[reg_season_total['Season'] >= 1995].copy()

# Create the plot
plt.figure(figsize=(10, 6))

# Regression line
sns.regplot(
    data=modern_data,
    x='3PA',
    y='ORtg',
    scatter=False,
    line_kws={'color': 'gray', 'alpha': 0.4, 'linewidth': 2}
)

# Scatter points sized by pace
sns.scatterplot(
    data=modern_data,
    x='3PA',
    y='ORtg',
    size='Pace',
    sizes=(40, 200),
    color='#1f77b4',
    legend=False
)

# Title and axis labels
plt.title("Regular Season: Offensive Rating vs 3-Point Attempts", fontsize=14, weight='bold')
plt.xlabel("3-Point Attempts", fontsize=12, weight='bold')
plt.ylabel("Offensive Rating", fontsize=12, weight='bold')

# Format x-axis ticks with commas
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

# Add bottom right note about size
plt.text(
    x=modern_data['3PA'].max() * 1.05, 
    y=modern_data['ORtg'].min() - 0.4,
    s="Note: Point size reflects pace (possessions per game)",
    fontsize=9,
    color='gray',
    ha='right'
)

# Clean styling
plt.tick_params(axis='both', labelsize=10)
sns.despine()

plt.tight_layout()

# Save the figure
plt.savefig("../visualizations/offensive_rating_v_3PA.png", dpi=300)

# Display the plot
plt.show()
```
