```python
# Load schedule and results
schedule_df = pd.read_csv(schedule_path)

# Load standings and game margins
standings_df = pd.read_csv(standings_path)

# Load league team stats (already cleaned)
team_stats_df = pd.read_csv(team_stats_path)

# Load cleaned team data
team_data_df = pd.read_csv(team_data_path)
```

```python
# Fix column names
schedule_df.columns = schedule_df.columns.str.strip().str.replace(" ", "_")
standings_df.columns = standings_df.columns.str.strip().str.replace(" ", "_")
team_stats_df.columns = team_stats_df.columns.str.strip().str.replace(" ", "_").str.replace("%", "Pct")
team_data_df.columns = team_data_df.columns.str.strip().str.replace(" ", "_").str.replace("%", "Pct")

# Fix column types
schedule_df["Date"] = pd.to_datetime(schedule_df["Date"], errors="coerce")
schedule_df["Season"] = schedule_df["Season"].astype(int)
standings_df["Season"] = standings_df["Season"].astype(int)
team_stats_df["SEASON"] = team_stats_df["SEASON"].astype(int)
team_data_df["Season"] = team_data_df["Season"].astype(int)

# Rename team_stats YEAR for merging
team_stats_df = team_stats_df.rename(columns={"SEASON": "Season"})

# Parse date
schedule_df["Date"] = pd.to_datetime(schedule_df["Date"], errors="coerce")
```

```python
# Models for different eras
sns.set_style("white")

# Define eras
era_bins = {
    "Early": (2000, 2010),
    "Middle": (2011, 2016),
    "Modern": (2017, 2024)
}

# Filter valid data
model_df = team_merged.dropna(subset=["WinPct", "3PPct", "AST", "TOV", "ORB", "eFGPct", "3PA", "2PA", "2PPct"])

# Store models
era_models = {}
coef_list = []  # List to store coefficients for plotting

# Fit model for each era
for era_name, (start, end) in era_bins.items():
    era_data = model_df[(model_df["Season"] >= start) & (model_df["Season"] <= end)].copy()
    X = era_data[["3PPct", "eFGPct", "AST", "TOV", "ORB", "3PA", "2PA", "2PPct"]]
    y = era_data["WinPct"]
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    era_models[era_name] = model
    print(f"\nRegression Summary: {era_name} Era ({start}–{end})")
    print(model.summary())
    
    # Collect coefficients and store in coef_list
    for var, coef in model.params.items():
        coef_list.append({
            "Era": era_name,
            "Variable": var,
            "Coefficient": coef
        })

# Create a DataFrame for the coefficients
coef_final = pd.DataFrame(coef_list)
```

### Key Trends Over Time:

#### Impact of Three-Point Shooting:
- In the **Early Era**, three-point shooting had little to no direct impact on win percentage.
- In the **Middle Era**, there was still a weak positive relationship with three-point shooting, though it wasn’t statistically significant.
- In the **Modern Era**, **3PPct** became a significant factor, underlining the growing importance of three-point shooting in determining success.

#### Importance of Assists:
- **AST** was a significant variable in the **Early Era**, and it remained significant throughout the periods, but it became less impactful in the **Modern Era** where other factors like three-point shooting became more dominant.

#### Turnovers and Offensive Rebounds:
- **TOV** consistently showed a negative and statistically significant relationship across all three eras, confirming that lower turnovers are crucial for winning.
- **ORB** remained positively significant across all periods, emphasizing the value of offensive rebounding.

#### Two-Point Attempts:
- In the **Early and Middle Eras**, more **2PA** was negatively correlated with win percentage, suggesting that shooting more two-pointers might have been less effective. This trend was not as pronounced in the **Modern Era**, where the game evolved towards perimeter shooting.

#### Conclusion:
Over time, the NBA has increasingly relied on the three-point shot, with its impact becoming more significant in recent years. Turnovers and offensive rebounds remain critical across all periods. The trend shows a shift in how teams are valuing shooting efficiency, particularly from beyond the arc, in the modern game.

```python
# Run a regression model

# Define the full list of features
features = ["Q('3PPct')", "Q('2PPct')", "Q('3PA')", "Q('2PA')", "AST", "TOV", "ORB", "eFGPct"]

# Ensure all features and target are non-null
regression_df = team_merged.dropna(subset=["WinPct", "3PPct", "AST", "TOV", "ORB", "3PA", "2PA", "2PPct", "eFGPct"]).copy()

# Build the formula string
formula = "WinPct ~ " + " + ".join(features)

# Fit the model
model = ols(formula, data=regression_df).fit()

# Print the regression summary
print(model.summary())
```

### Summary:

- **R-squared**: 0.479 (47.9% of the variation in **WinPct** is explained by the model). This suggests the model is a moderate fit.

#### Significant Predictors:
- **2PPct (Two-Point Percentage)**: Negative and significant (p = 0.008), indicating that higher two-point shooting percentage is associated with a lower win percentage.
- **3PA (Three-Point Attempts)**: Negative and significant (p < 0.001), suggesting that more three-point attempts negatively impact win percentage.
- **2PA (Two-Point Attempts)**: Negative and significant (p < 0.001), indicating that more two-point attempts are detrimental to win percentage.
- **AST (Assists)**: Positive and significant (p < 0.001), showing that more assists are associated with a higher win percentage.
- **TOV (Turnovers)**: Negative and significant (p < 0.001), indicating that fewer turnovers lead to a higher win percentage.
- **ORB (Offensive Rebounds)**: Positive and significant (p < 0.001), showing that more offensive rebounds contribute to a higher win percentage.
- **eFGPct (Effective Field Goal Percentage)**: Positive and significant (p < 0.001), indicating that higher shooting efficiency correlates with a higher win percentage.

#### Non-Significant Variables:
- **3PPct (Three-Point Percentage)**: Not significant (p = 0.628), indicating that three-point shooting percentage did not have a direct impact on win percentage in this model.

```python
# Clustering

# Feature Engineering: Add 3P rate and eFG% if not present
team_data_df["3P_Rate"] = team_data_df["3PA"] / team_data_df["FGA"]
team_data_df["eFGPct"] = (team_data_df["FG"] + 0.5 * team_data_df["3P"]) / team_data_df["FGA"]

# Select relevant features
features = ["3P_Rate", "3PPct", "eFGPct", "WinPct"]
cluster_df = team_data_df[features].dropna().copy()

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(cluster_df)

# Fit KMeans clustering model
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_df["Cluster"] = kmeans.fit_predict(X_scaled)

# ✅ Add cluster labels back to team_data_df
team_data_df.loc[cluster_df.index, "Cluster"] = cluster_df["Cluster"]

# Define cluster labels (based on cluster_summary insight)
labels = {
    0: "Low 3P, Low Win%",
    1: "Efficient 3P, High Win%",
    2: "High 3P Volume, Good Win%"
}

# Map descriptive labels
cluster_df["Style"] = cluster_df["Cluster"].map(labels)

# PCA for visualization (2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
cluster_df["PC1"] = X_pca[:, 0]
cluster_df["PC2"] = X_pca[:, 1]

# Plot with Descriptive Labels
plt.figure(figsize=(10, 6))
sns.set_style("white")
sns.scatterplot(
    data=cluster_df,
    x="PC1",
    y="PC2",
    hue="Style",
    palette="Set2",
    s=80,
    edgecolor="black"
)
plt.title("Team Clusters Based on 3P Reliance and Efficiency", fontsize=14, weight="bold")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Playing Style", bbox_to_anchor=(1.02, 1), loc="upper left")
sns.despine()
plt.tight_layout()
plt.savefig("../visualizations/team_clusters_labeled.png", dpi=300)
plt.show()

# Summarize Clusters with Labels
cluster_summary = cluster_df.groupby("Style")[features].mean().round(3)
display(cluster_summary)
```
